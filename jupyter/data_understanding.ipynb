```python
class insuranceHandler():
    def __init__(self, path, data_name, 
                       insurance_code_list= {'coramo': ['0977']}, 
                       parse_dates=['damesano'], 
                       fields_type={'damesano':str, 
                                    'coramo':str}):
        
        date_format = lambda x: pd.datetime.strptime(x, '%Y%m')
        in_file = os.path.join(path, data_name)
        
        iter_csv = pd.read_csv(in_file, 
                              parse_dates=parse_dates, 
                              date_parser=date_format,
                              encoding='latin-1', 
                              sep=';',
                              dtype = fields_type,
                              decimal=',',
                              iterator = True,
                              chunksize = 1000)
        
        filter_var = list(insurance_code_list.keys())[0]
        filter_ = insurance_code_list.get(filter_var)
        
        self.df = pd.concat([chunk[chunk[filter_var].isin(filter_)] for chunk in iter_csv])
        
    def apply_formula(self, formula):
        var_, formula_ = formula.replace(' ', '')split('=')
        self.df[var_] = self.df.eval(formula_)
        
```

```python
fields_prov = {'damesano':str , 
               'coenti':np.int, 
               'coramo':str, 
               'ppng':np.float,  
               'ppngretro':np.float,  
               'ppngrvne':np.float,  
               'pip':np.float, 
               'psl':np.float,  
               'ibnr':np.float,  
               'ibner':np.float,  
               'pda':np.float, 
               'pcp':np.float,  
               'outras':np.float,  
               'provbenefconceder':np.float, 
               'provbenefconcedido':np.float, 
               'rec_sinliq':np.float, 
               'rec_sinliqadm':np.float, 
               'rec_sinliqjud':np.float, 
               'rec_ibnr':np.float,  
               'rec_ibner':np.float,  
               'pdr':np.float, 
               'pcc':np.float}
```

```python
def handle_field_type(df, dict_type, prod_list):
    df_ = df[dict_type.keys()].copy()
    for field, type_ in dict_type.items():
        df_[field] = df_[field].apply(lambda x: str(x).replace(',','.'))
        df_[field] = df_[field].fillna('0')
        df_[field] = df_[field].astype(type_)
        
    df_['damesano_dt'] = pd.to_datetime(df_['damesano'], format="%Y%m")
    df_ = df_[df_['coramo'].isin(prod_list)]
    return df_

#prod_list = ['0983', '0986', '1383', '1386']
#prod_list = ['0991', '0993', '1391']
#prod_list = ['0992', '0994', '1392'] + ['0983', '0986', '1383', '1386']
prod_list = ['0977']

ses_insurance_df = handle_field_type(ses_insurance_df, fields, prod_list)
ses_insurance_prov_df = handle_field_type(ses_insurance_prov_df, fields_prov, prod_list)
```
